```python
import pandas as pd
import numpy as np
import keras
import tensorflow
from keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Embedding, Flatten, Dense
```

```python
df=pd.read_csv ('/content/drive/MyDrive/dataset/emotions.csv')
df.head(10)
```

```python
texts = df["text"].tolist()
labels = df["label"].tolist()
print(labels)
```

**Tokanize**

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)
```

```python
sequences = tokenizer.texts_to_sequences(texts)
max_length = max([len(seq) for seq in sequences])
padded_sequences = pad_sequences(sequences, maxlen=max_length)
```

**Split** the Dataset


```python
X_train, X_test, y_train, y_test = train_test_split(padded_sequences, labels, test_size=0.2, random_state=42)
```

**Define LSTM**

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense
from keras.utils import to_categorical


embedding_size = 128
lstm_units = 64

model = Sequential()
model.add(Embedding(input_dim=len(tokenizer.word_index) + 1, output_dim=embedding_size, input_length=max_length))
model.add(LSTM(units=lstm_units))
model.add(Dense(6, activation='softmax'))

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

**Train The Model**

```python
batch_size = 32
epochs = 10
model.fit(X_train, np.array(y_train), batch_size=batch_size, epochs=epochs, validation_split=0.1)
```

**Accuracy**

```python
loss, accuracy = model.evaluate(X_test, np.array(y_test))
print(f'Test Loss: {loss}, Test Accuracy: {accuracy}')
```

**Predict**

```python
import random

# Select a random sentence from the dataset
random_index = random.randint(0, len(df) - 1)
random_sentence = df.loc[random_index, 'text']

# Predict the label for the randomly selected sentence
sequence = tokenizer.texts_to_sequences([random_sentence])
padded_sequence = pad_sequences(sequence, maxlen=max_length)
predicted_probabilities = model.predict(padded_sequence)
predicted_label_index = np.argmax(predicted_probabilities)
predicted_emotion = ['sadness', 'joy', 'love', 'anger', 'fear', 'surprise'][predicted_label_index]

print(f"Randomly Selected Sentence: {random_sentence}")
print(f"Predicted Emotion: {predicted_emotion}")

```
